# Data Preprocessing

```python
import pandas as pd
df = pd.read_csv(r"../Kickstarer_dataset/kickstarter_combined.csv")
df.head(3)
```

```python
# Check the no. of items
len(df)
```

```python
df = df.drop_duplicates()
```

```python
import json
```

```python
df['category'] = pd.Series([json.loads(df['category'][x]) for x in range(len(df['category'])-1)])
df['category'] = pd.Series([df['category'][x]['name'] for x in range(len(df['category'])-1)])

df['urls'] = pd.Series([json.loads(df['urls'][x]) for x in range(len(df['urls'])-1)])
df['urls'] = pd.Series([df['urls'][x]['web']['project'] for x in range(len(df['urls'])-1)])

df.head(3)
```

```python
df['usd_goal'] = df['fx_rate'] * df['goal']
```

```python
df = df.drop(columns=['goal', 'country_displayable_name', 'currency', 'currency_symbol', 'friends', 'fx_rate', 'static_usd_rate', 'usd_exchange_rate', 'usd_type', 'creator', 'currency_trailing_code', 'disable_communication', 'id', 'is_backing', 'is_starrable', 'is_starred'])
df.head(3)
```

```python
df = df.drop(columns=['converted_pledged_amount', 'pledged', 'location', 'permissions', 'profile', 'slug', 'source_url', 'photo', 'current_currency'])
df.head(3)
```

```python
# Convert timestamp to datetime
df['created_at'] = pd.to_datetime(df['created_at'], unit="s")
df['deadline'] = pd.to_datetime(df['deadline'], unit="s")
df['launched_at'] = pd.to_datetime(df['launched_at'], unit="s")
df['state_changed_at'] = pd.to_datetime(df['state_changed_at'], unit="s")
df.head(3)
```

```python
df = df.drop(columns=['created_at', 'state_changed_at'])
df.head(3)
```

```python
# pledged ratio
df['pledged_ratio'] = (df['usd_pledged'] / df['usd_goal'] * 100).astype(int)
df.head(3)
```

```python
# sort by
df = df.sort_values(by='usd_pledged', ascending=False)
df.head()
```

```python
# rearrange the columns
df = df[['category', 'launched_at', 'state', 'name', 'blurb', 'country', 'spotlight', 'staff_pick', 'backers_count', 'usd_goal', 'usd_pledged', 'pledged_ratio', 'urls']]
df.head(3)
```

```python
df = df.drop_duplicates(subset='name', keep='last')
len(df)
```

# 2. By Keywords

```python
search_list = ['cup', 'mug', 'tumbler', 'bottle', 'coffee cup', 'coffee mug', 'coffee tumbler']
```

```python
df_by_keyword = df
```

```python
# Option 1 - Search the names
df_by_keyword = df_by_keyword.loc[df_by_keyword.name.str.contains('|'.join(search_list), na=False, regex=True)]
len(df_by_keyword)
```

```python
# Option 2 - Search the blurbs
import re
df_by_keyword = df_by_keyword.loc[df_by_keyword.blurb.str.contains('|'.join(search_list), na=False, regex=True, flags=re.IGNORECASE)]
len(df_by_keyword)
```

```python
# Drop duplicates
df_by_keyword = df_by_keyword.drop_duplicates(subset='name', keep='last')
len(df_by_keyword)
```

```python
df_by_keyword = df_by_keyword.sort_values('category')
list(df_by_keyword['category'].unique())
```

```python
# Delete unnecessary categories
cate_to_select = [
    #'Accessories',
    'Ceramics',
    #'Crafts',
    #'DIY',
    'Design',
    #'Drinks',
    #'Food',
    #'Food Trucks',
    'Gadgets',
    'Glass',
    'Hardware',
    'Interactive Design',
    'Pottery',
    'Print',
    'Printing',
    'Product Design',
    #'Restaurants',
    #'Robots',
    #'Small Batch',
    'Stationery',
    'Technology',
    'Woodworking',
    'Young Adult']
```

```python
df_by_keyword = df_by_keyword.loc[df_by_keyword.category.str.contains('|'.join(r'^{cate_to_select}'), case=True, regex=True)]
len(df_by_keyword)
```

#Word Cloud

```python
import numpy as np # linear algebra
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS
```

```python
mpl.rcParams['figure.figsize']=(12.0,9.0)    #(6.0,4.0)
mpl.rcParams['font.size']=12                #10 
mpl.rcParams['savefig.dpi']=100             #72 
mpl.rcParams['figure.subplot.bottom']=.1 


stopwords = set(STOPWORDS)
data = pd.read_csv(path)
```

```python
wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(data['name']))

print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("word1.png", dpi=900)
```

```python
wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(data['blurb']))

print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("word1.png", dpi=900)
```

```python
wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=60, 
                          random_state=84
                         ).generate(str(data['name' and 'blurb']))

print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("word1.png", dpi=900)
```
